```python
import cv2
import mediapipe as mp
import numpy as np
import os
import csv
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
import pickle
```

```python
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose
```

# Extraction des coordonnées

```python
directory =r'C:\Users\Sohaib\OneDrive - UMONS\Documents\Projects\DL4HAR_StudentsProj\Projet_NTU\data\nturgb+d_rgb_videos'
#répertoire avec toutes les vidéos nturgb+d
```

```python
all_videos_names = os.listdir(directory)
#print(all_videos_names)
#len(all_videos_names) #au total : 56881 vidéos
#on vient récupérer tous les noms de vidéos du répertoire dans une lsite all_videos_names
```

```python
index = 0
all_data = {}
```

```python
#On va boucler parmis toutes les vidéos pour récupérer les coordonnées
for counter in enumerate(all_videos_names, 0):
        
    vidlist = []

    # Retrieve a Class Name using the Random Index.
    selected_video_file_name = all_videos_names[index]
    
    #The two second line allows to get the name of the video without the extension
    file_path = os.path.splitext(f'{directory}/{selected_video_file_name}')[0]
    file_name = file_path.split('/')[-1]
    
    if 'rgb.avi' not in selected_video_file_name:
        print('skipping', selected_video_file_name)
    
    elif file_name not in all_data.keys() and int(file_name[-7:-4]) <= 49:
        # Initialize a VideoCapture object to read from the video File. 
        cap = cv2.VideoCapture(f'{directory}/{selected_video_file_name}')
        
        with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:

            while cap.isOpened():
                success, image = cap.read()

                if not success:
                    # print("Ignoring empty camera frame.")
                    # If loading a video, use 'break' instead of 'continue'.
                    break

                # To improve performance, optionally mark the image as not writeable to
                # pass by reference.

                image.flags.writeable = False
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                results = holistic.process(image)


                # Extract landmarks
                try:
                    landmarks = results.pose_world_landmarks.landmark #fonction qui permet d'avoir les coordonnées avec comme origine le centre des hanches
                except:
                    pass


            # Draw landmark annotation on the image.

                image.flags.writeable = True
                image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
                mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())



             # Export coordinates
                try:

                # Extract Pose landmarks

                    pose = results.pose_world_landmarks.landmark
                    pose_row = np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose])
                    vidlist.append(pose_row)

                except:
                    pass
                    # cv2.imshow('Raw Webcam Feed', image)
                    if cv2.waitKey(10) & 0xFF == ord('q'):
                        break
        
        #mp_drawing.plot_landmarks(results.pose_world_landmarks,mp_pose.POSE_CONNECTIONS)
        all_data[file_name] = np.array(vidlist)
        
        cap.release()
        cv2.destroyAllWindows()
    else:
        print('skipping file', file_name)
    index += 1
    print('{}/{}'.format(index, len(all_videos_names)))
    
#np.save('1erdictionnaire.npy', all_data)
        

```

```python
np.save('1erdictionnaire.npy', all_data)
```
